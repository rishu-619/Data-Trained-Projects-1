```python
x=df.iloc[:,1:8]
y=df.iloc[:,-1]
```

```python
def linreg():
    lm=LinearRegression()
    lm.fit(x,y)
    print('score=',lm.score(x,y))
    print('coefficient=',lm.coef_)
    print('Intercept=',lm.intercept_)
```

```python
linreg()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=0)
```
